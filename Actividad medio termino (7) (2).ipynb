###### En este caso me enfocar√© en las variables de Total cases, Total deaths,Total recovered, Country and Total Tests.

```python
df.drop(['Tot Cases/ 1M pop','Deaths/ 1M pop','Active Cases','Tests/ 1M pop','Population',],
        axis= 1,
        inplace= True)
df.columns
```

## Variable Description

#### Country-  It's a categorical variable that shows the names of the countries
#### Total Cases- It's a numerical variable that shows the total of cases on Oceania
#### Total deaths - It's a numerical variable that shows the total of deaths on Oceania
#### Total recovered - It's a numerical variable that shows the total of pacients who recovered from covid-19
#### Total Tests - It's a numerical variable that shows the total number of tests taken

## Questions

#### 1. What is the country with the highest number of cases on the continent of Oceania?

```python
myPivot = pd.pivot_table(df,
                         index = ['Country'],
                         values= ['Total Cases'],
                        )
myPivot
```

```python
import matplotlib.pyplot as plt
```

```python
my_plot = df.plot("Country", "Total Cases", kind="bar")
plt.show() 
```

#### In conclusion we can say that Australia far exceeds other countries in terms of the number of covid-19 cases

#### 2. Wich is the country with the fewest deaths on the continent of Oceania?

```python
myPivot = pd.pivot_table(df,
                         index = ['Country'],
                         values= ['Total Deaths'],
                         aggfunc=['sum'],
                        )
myPivot
```

```python
my_plot = df.plot("Country", "Total Deaths", kind="barh")
plt.show() 
```

#### En conclusi√≥n podemos decir que Palau es el pa√≠s que cuenta con la menor cantidad de muertes por covid-19.

#### 3. Cual es el pa√≠s que cuenta con mayor cantidad de pacientes recuperados de covid-19?

```python
myPivot = pd.pivot_table(df,
                         index = ['Country'],
                         values= ['Total Recovered'],
                         aggfunc=['sum'],
                        )
myPivot
```

```python
my_plot = df.plot("Country", "Total Recovered", kind="bar")
plt.show() 
```

#### En conclusi√≥n podemos decir que Australia es el pa√≠s que cuenta con m√°s pacientes recuperados de Covid-19
