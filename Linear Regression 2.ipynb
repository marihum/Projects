```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
import statsmodels.api as sm
```

```python
df = pd.read_csv('antropometria-dataset.csv')
df
```

```python
df.dropna(inplace = True)
df
```

```python
y = pd.DataFrame(df['cadera'])
x = pd.DataFrame(df['peso'])

lm = linear_model.LinearRegression()
model_lm = lm.fit(x, y)
model_lm
```

```python
print('The y-intercep, b0 = ', model_lm.intercept_)
print('The coefficient of peso, b1 = ', model_lm.coef_)
print('The R-squared value is = ', model_lm.score(x,y))
```
